# 03 - Pré-processamento

Cruzamento CNEFE + Web Scraping via DuckDB, seleção de features e normalização.

O resultado final é exportado como parquet para uso na modelagem.

```python
import duckdb
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from pathlib import Path

con = duckdb.connect()
CNEFE   = "'../data/cnefe_2010/*.snappy.parquet'"
SCRAPING = "'../data/web_scraping_pre_processado/*.snappy.parquet'"
OUTPUT_DIR = Path('../outputs')
```

```python
# Cruzamento pelo CEP — ajuste o critério de join conforme os dados
df = con.execute(f"""
    SELECT
        c.uf,
        c.munic,
        c.setor,
        c.tipo,
        c.situacao,
        s.WEBSCRAPING_TIPO_ESTABELECIMENTO
    FROM read_parquet({CNEFE}) c
    INNER JOIN read_parquet({SCRAPING}) s
        ON c.cep = s.WEBSCRAPING_CEP
    LIMIT 1000000
""").df()

print(df.shape)
df.head()
```
